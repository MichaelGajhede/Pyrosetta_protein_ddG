```python
import pandas as pd
import seaborn
import matplotlib
```

```python
df = pd.read_csv ('1BV1_mut_old.csv')
print (df)
```

```python
df.nlargest(5, ['ddG'])
```

```python
df.nsmallest(200, ['ddG'])
```

```python
df_pivot = df.pivot("resi", "resn", "ddG")
df_pivot
```

```python
df_mins = df_pivot.min(axis=1) #returns panda series which is an indexed 1d array
df_mins
```

```python
from Bio import *
from Bio.PDB import PDBParser, PDBIO
```

```python
io = PDBIO()
p = PDBParser()

structure = p.get_structure("1bv1", "1bv1_copy.pdb") # input files

for model in structure:
    for chain in model:
        for residue in chain:
            resi = residue.get_id()[1]
            #print(resi)
            bfac = df_mins.loc[resi]
            #print(bfac)
            for atom in residue:  
                atom.set_bfactor(bfac) # set b to ddG min for residue

print(structure)
io.set_structure(structure)
io.save("pymol-out.pdb")
                
```
